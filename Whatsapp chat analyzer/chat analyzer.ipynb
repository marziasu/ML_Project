### this text file convert into pandas dataframe

```python
# regular expression for message extract
pattern=r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s*[AP]M\s-\s'
```

```python
messages=re.split(pattern, data)[1:]
messages
```

```python
dates=re.findall(pattern, data)
dates
```

```python
df=pd.DataFrame({
    'user_message': messages,
    'message_date': dates
})
```

```python
df
```

```python
df['message_date'].sample(5)
```

```python
# convert time data to datetime format
df['message_date']= pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p - ' ) # format is for 12 hour time
df.rename(columns={'message_date': 'date_time'}, inplace=True )
df.head()
```

```python
df.shape
```

```python
# separate users and message
users=[]
messages=[]

for message in df['user_message']:
    entry= re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group notification')
        messages.append(entry[0])
df['user']= users
df['message']= messages

df.drop(columns=['user_message'], inplace=True )
df.head()
```

```python
df['year']= df['date_time'].dt.year
df['month']= df['date_time'].dt.month_name()
df['day']= df['date_time'].dt.day
df['hour']= df['date_time'].dt.hour
df['minute']= df['date_time'].dt.minute
```

### Monthly Timeline

```python
# plot in which message in x-axis and time(month+year) in y-axis
# find total message according to year (using groupby)
df['month_num']=df['date_time'].dt.month    # convert month_name to month_num so that easily sort month for msg count
df.head()
```

```python
monthly_timeline= df.groupby(['year', 'month_num','month']).count()['message'].reset_index()
monthly_timeline
```

```python
# merge year+month
time=[]
for i in range(monthly_timeline.shape[0]):
    time.append(monthly_timeline['month'][i]+'-'+str(monthly_timeline['year'][i]) )
time
```

```python
monthly_timeline['time']=time
monthly_timeline
```

```python
plt.plot(monthly_timeline['time'], monthly_timeline['message'] )
plt.xticks( rotation= 'vertical' )
plt.show()
```

### heatmap, busy time in week

```python
# hour in x-axis , day_name in y_axis
df['time']= df['date_time'].dt.time
df['time']
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour==23:
        period.append(str(hour)+'-'+str('00'))
    elif hour==0:
        period.append(str('00')+'-'+str(hour+1) )
    else:
        period.append(str(hour)+'-'+str(hour+1) )
df['period']=period
df.head()
```

```python
sns.heatmap( df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count' ).fillna(0) )
```
